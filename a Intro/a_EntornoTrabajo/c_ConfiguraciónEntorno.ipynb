## Terminal

**The terminal** is a text-based interface that allows users to interact with their computer's operating system by typing commands. It provides direct access to the systemâ€™s shell, which interprets and executes commands. The terminal is commonly used for tasks like navigating the file system, running scripts, managing processes, and configuring system settings.

Key Features:
* **Command-Line Interface (CLI):** Unlike graphical user interfaces (GUIs), the terminal operates entirely through text commands.


* **Shell Integration:** The terminal runs a shell (e.g., Bash, Zsh, PowerShell) that processes user commands.


* **Automation:** Commands can be scripted for automation using shell scripts.


* **Remote Access:** It can be used to connect to remote servers via SSH.


* **Efficiency:** Many tasks can be performed faster with commands compared to GUI-based interactions.



The following shells depend on the operating system:

* **Windows, (Linux, Mac):** [PowerShell][PS], [cheatsheet][CSHT]


* **Mac / Linux:** [Bash][BSH], [zsh][ZSH]

### Setting up the working python-environment

1. To create an environment called ```nosql``` with which we will work in the course.
```terminal
> conda create --name nosql -y
```

2. Activate the ```nosql``` environment
```terminal
> conda activate nosql
```

* To install additional libraries
```terminal
> conda install pandas matplotlib numpy -y
```

* To delete an environment called ```nosql```
```terminal
> conda env remove -n nosql
```

* To deactivate an environment
```terminal
> conda deactivate
```

### Installing *Jupyter Notebooks* and adding the ```nosql``` kernel

* Install *Jupyter Notebooks*
```terminal
> conda install jupyter -y
```

* Show the list of installed kernels in Jupyter Notebooks
```terminal
> jupyter kernelspec list
```

* We install the ```ipykernel``` library, which allows adding kernels to Jupyter Notebooks
```terminal
> conda install ipykernel -y
```

* We register the ```nosql``` environment (kernel) in Jupyter Notebooks.
```terminal
> python -m ipykernel install --user --name=nosql
```

* We verify that the 'nosql' kernel can be changed from the menu "Kernel-Change Kernel" to ```nosql```

## Jupiter Notebooks

[**Jupyter Notebooks**][NB] is an open-source, interactive web application that allows you to create and share documents that contain live code, equations, visualizations, and narrative text. It is commonly used for data science, machine learning, and academic research.

Some key features of Jupyter Notebooks are:

* **Interactive Code:** You can write and execute code in various programming languages (such as Python, R, Julia, etc.) directly within the notebook.


* **Data Visualization:** It supports visualizations with libraries like matplotlib, seaborn, and plotly, enabling you to create interactive and static plots.


* **Markdown and Rich Text:** You can use Markdown to add descriptive text, including formatting, equations (using LaTeX), and images.


* **Reproducible Research:** Jupyter Notebooks are great for documenting your code, results, and methodology, making it easier to share and reproduce research.


* **Flexible Kernels:**You can use different programming languages (kernels) within a single notebook, depending on the installed kernels.

## Markdown

**Markdown** is a lightweight markup language that allows you to format text using plain text syntax. It is designed to be easy to read and write while providing a simple way to apply formatting like headings, bold text, lists, and links. Markdown is widely used for documentation, readme files, blogging, and in Jupyter Notebooks for adding rich text content alongside code.

Here are some basic elements of Markdown:

- **Headings:** Use # symbols to create headings.
```markdown
# Heading 1
## Heading 2
### Heading 3
```

- **Bold Text:** Wrap the text in double asterisks ```**``` or double underscores ```__```

```markdown
**Bold Text** or __Bold Text__
```

- **Italic Text:** Wrap the text in single asterisks ```*``` or single underscores```_```

```markdown
*Italic Text* or _Italic Text_
```

- **Lists:**

- **Unordered** list: Use *, -, or + followed by a space.

```markdown
* Item 1
- Item 2
```

- **Ordered** list: Use numbers followed by a period.

```markdown
1. First item
2. Second item
```

- **Links:** Use square brackets for the link text and parentheses for the URL.

```markdown
[OpenAI](https://www.openai.com)
```

- **Images:** Similar to links, but with an exclamation mark at the beginning.

```markdown
![Image Alt Text](image_url)
```

- **Code:** Inline code is wrapped in backticks `code` and code blocks are wrapped in triple backticks.
- Inline: 

```markdown
`print("Hello")`
```

-Block:

```markdown
def greet():
    print("Hello World")
```

Blockquotes: Use the > symbol to create a blockquote.
```markdown
> This is a quote.
```
