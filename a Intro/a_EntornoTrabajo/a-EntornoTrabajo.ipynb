## Configuración de *Miniconda*

### 1. Descargar e instalar Miniconda  
- Descargar el instalador `.exe` de Miniconda desde el siguiente enlace: [Presionar aquí][MINI].  
   
  [MINI]: <https://repo.anaconda.com/miniconda/>  

- Ejecutar el archivo `.exe` descargado y seguir las instrucciones en pantalla.  
- Asegurarse de seleccionar la opción **"Add Miniconda to my PATH environment variable"** durante la instalación para facilitar el uso de `conda` desde la terminal.  

### 2. Verificar la instalación  
- Abrir la **línea de comandos** (CMD) o **Anaconda Prompt**.  
- Escribir el siguiente comando para comprobar que Miniconda se ha instalado correctamente:  

  ```bash
  conda --version

### Crear un nuevo ambiente
- Para crear un nuevo ambiente virtual, ejecutar:
conda create -n mi_entorno python=3.10

### Activar el ambiente
#### Windows:
- conda activate mi_entorno

#### Linux/macOs:
- source activate mi_entorno

### Instalar paquetes
- Instalar paquetes dentro del ambiente:
conda install numpy pandas matplotlib
- Si se necesita instalar paquetes con pip:

pip install requests

### Desactivar y eliminar un ambiente
- Para desactivar el ambiente actual:
conda deactivate

## Resumen de los elementos principales de Markdown

Markdown es un lenguaje de marcado ligero que permite dar formato a textos de manera sencilla. Sus principales características incluyen:

- **Encabezados**: Se crean con `#`, `##`, `###`, etc., según el nivel.
- **Énfasis**: Se usa `*` o `_` para cursiva y `**` o `__` para negrita.
- **Listas**: Se crean con `-` o `*` para listas desordenadas y con números para listas ordenadas.
- **Enlaces e imágenes**: Se definen con `[texto](url)` para enlaces y `![alt](url)` para imágenes.
- **Código**: Se puede insertar código en línea con `` `código` `` o en bloque con triple tilde (`` ``` ``).
- **Citas**: Se indican con `>`.
- **Tablas**: Se construyen con `|` y `---` para definir filas y columnas.
- **Líneas horizontales**: Se generan con `---`.

Markdown es ampliamente usado en documentación, blogs y plataformas como GitHub debido a su simplicidad y compatibilidad con múltiples herramientas.

## Bases de Datos

En este curso, trabajaremos con diferentes sistemas de gestión de bases de datos (DBMS), cada uno con sus herramientas asociadas. A continuación, se describen sus características, ventajas y ejemplos de uso.

### 1. SQLite
- **Descripción**: SQLite es una base de datos ligera y de código abierto que no requiere configuración de servidor. Es ampliamente utilizada para aplicaciones móviles, desarrollo rápido de prototipos y almacenamiento local en navegadores.
- **Ventajas**:
  - No requiere instalación de servidor.
  - Es eficiente y de bajo consumo de recursos.
  - Compatible con múltiples lenguajes de programación.
- **Ejemplo de uso**:
  ```sql
  CREATE TABLE usuarios (
      id INTEGER PRIMARY KEY,
      nombre TEXT NOT NULL,
      edad INTEGER
  );
  INSERT INTO usuarios (nombre, edad) VALUES ('Juan', 30);
  SELECT * FROM usuarios;
  ```
- **IDE recomendado**: [DB Browser for SQLite](https://sqlitebrowser.org/)

### 2. PostgreSQL
- **Descripción**: PostgreSQL es un sistema de gestión de bases de datos relacional avanzado, de código abierto y altamente extensible. Se usa en entornos empresariales y proyectos que requieren integridad de datos y escalabilidad.
- **Ventajas**:
  - Soporta estructuras de datos avanzadas como JSON, XML y arrays.
  - Ofrece robustez y estabilidad en transacciones.
  - Soporta replicación y alta disponibilidad.
- **Ejemplo de uso**:
  ```sql
  CREATE TABLE empleados (
      id SERIAL PRIMARY KEY,
      nombre VARCHAR(100) NOT NULL,
      salario DECIMAL(10,2)
  );
  INSERT INTO empleados (nombre, salario) VALUES ('María', 50000.00);
  SELECT * FROM empleados;
  ```
- **IDE recomendado**: [pgAdmin](https://www.pgadmin.org/)

### 3. MongoDB
- **Descripción**: MongoDB es una base de datos NoSQL orientada a documentos, que almacena los datos en formato JSON flexible. Se utiliza en aplicaciones que requieren almacenamiento escalable y esquemas dinámicos.
- **Ventajas**:
  - Permite almacenamiento flexible sin esquema fijo.
  - Ofrece alta escalabilidad horizontal.
  - Soporta consultas avanzadas con agregaciones.
- **Ejemplo de uso**:
  ```json
  {
      "nombre": "Carlos",
      "edad": 28,
      "direccion": {
          "ciudad": "Madrid",
          "pais": "España"
      }
  }
  ```
- **Herramientas recomendadas**:
  - [MongoDB Compass](https://www.mongodb.com/products/compass) (GUI oficial)
  - [Mongo Shell](https://www.mongodb.com/try/download/shell) (Interfaz de línea de comandos)

### 4. Neo4j
- **Descripción**: Neo4j es una base de datos orientada a grafos que permite modelar y consultar relaciones complejas de manera eficiente. Se usa en análisis de redes sociales, detección de fraudes y sistemas de recomendaciones.
- **Ventajas**:
  - Optimizado para consultas de relaciones complejas.
  - Soporta Cypher, un lenguaje de consulta específico para grafos.
  - Permite consultas en tiempo real sobre estructuras de datos interconectadas.
- **Ejemplo de uso**:
  ```cypher
  CREATE (persona:Usuario {nombre: 'Lucía', edad: 25})
  CREATE (amigo:Usuario {nombre: 'Pedro', edad: 27})
  CREATE (persona)-[:AMIGO_DE]->(amigo);
  MATCH (n) RETURN n;
  ```
- **Herramientas recomendadas**:
  - [Neo4j Desktop](https://neo4j.com/download/) (Aplicación de escritorio)
  - [Neo4j Browser](https://neo4j.com/developer/neo4j-browser/) (Interfaz web para interactuar con la base de datos)

Con estos sistemas de bases de datos, tendrás una base sólida para trabajar en proyectos que requieran almacenamiento eficiente y flexible según los diferentes casos de uso.
