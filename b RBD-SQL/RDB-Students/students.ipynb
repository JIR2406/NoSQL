## Importaci√≥n de los datos CSV a una base de datos SQLite



1. **Importar las librer√≠as necesarias**:
    ```python
    import pandas as pd
    ```

2. **Cargar el archivo CSV**:
    ```python
    csv_file = "students.csv"
    ```

    ```python

    ```

6. **Insertar los datos en la tabla**:
    ```python
    df.to_sql("Students", conn, if_exists="append", index=False)
    ```

7. **Confirmar y cerrar conexi√≥n**:
    ```python
    conn.commit()
    ```

````python
import sqlite3
import pandas as pd

# 1Ô∏è‚É£ Cargar el archivo CSV
csv_file = "students.csv"
df = pd.read_csv(csv_file)

# 2Ô∏è‚É£ Conectar a la base de datos SQLite
conn = sqlite3.connect("students.db")
cursor = conn.cursor()

# 3Ô∏è‚É£ Generar la sentencia CREATE TABLE autom√°ticamente
column_types = {
    "int64": "INTEGER",
    "float64": "REAL",
    "object": "TEXT"
}

columns_sql = ", ".join([f"{col} {column_types[str(dtype)]}" for col, dtype in df.dtypes.items()])
create_table_sql = f"CREATE TABLE IF NOT EXISTS Students ({columns_sql})"

# 4Ô∏è‚É£ Ejecutar la creaci√≥n de la tabla
cursor.execute(create_table_sql)

# 5Ô∏è‚É£ Insertar los datos en la tabla
df.to_sql("Students", conn, if_exists="append", index=False)

# 6Ô∏è‚É£ Confirmar y cerrar conexi√≥n
conn.commit()
conn.close()

print("Base de datos creada autom√°ticamente y datos importados.")
````

# Proceso de normalizaci√≥n de la base de datos
En el proceso de normalizaci√≥n de una base de datos, se busca organizar los datos de manera que se eliminen redundancias y dependencias innecesarias, dividiendo las tablas en estructuras m√°s peque√±as y f√°ciles de mantener. Este proceso se realiza en varias formas normales, siendo las tres primeras las m√°s comunes. A continuaci√≥n, te explico brevemente cada una:

### 1. Primera Forma Normal (1FN)
Una tabla est√° en **Primera Forma Normal** si:
- Todos los valores son at√≥micos (es decir, no hay grupos repetitivos de columnas o valores m√∫ltiples en una misma celda).
- Cada columna contiene un √∫nico valor.
- Cada fila es √∫nica.

### 2. Segunda Forma Normal (2FN)
Una tabla est√° en **Segunda Forma Normal** si:
- Cumple con la Primera Forma Normal.
- Todas las columnas no clave dependen completamente de la clave primaria, es decir, no debe haber dependencias parciales.

### 3. Tercera Forma Normal (3FN)
Una tabla est√° en **Tercera Forma Normal** si:
- Cumple con la Segunda Forma Normal.
- No existen dependencias transitivas, es decir, las columnas no clave no deben depender de otras columnas no clave.

---

## An√°lisis de la Tabla `Students`

Analizando la tabla `Students` que contiene informaci√≥n sobre los estudiantes, como su nombre, departamento, GPA, edad y a√±o de graduaci√≥n, podemos verificar en qu√© forma normal se encuentra:

### 1. Cumple con la Primera Forma Normal (1FN)
La tabla **cumple con la Primera Forma Normal** porque:
- Todos los valores en las celdas son at√≥micos.
- No existen columnas con m√°s de un valor (como una lista de valores).
- Cada fila es √∫nica.

### 2. Cumple con la Segunda Forma Normal (2FN)
La tabla **cumple con la Segunda Forma Normal** porque:
- Todas las columnas no clave dependen completamente de la clave primaria. 
- En este caso, la clave primaria es el `StudentID` y todos los atributos (como el nombre, GPA, edad y departamento) dependen directamente de este ID. No existen dependencias parciales.

### 3. No cumple con la Tercera Forma Normal (3FN)
La tabla **no cumple con la Tercera Forma Normal** debido a:
- La columna `DepartmentID` depende de otra columna no clave.
- Esto genera una dependencia transitiva, ya que `DepartmentID` puede estar asociado con informaci√≥n adicional del departamento (como su nombre), pero esta informaci√≥n se encuentra en la misma tabla, lo que genera redundancia y hace que no se cumpla con la 3FN.

---
## Propuesta de Modificaci√≥n para Cumplir con la Tercera Forma Normal (3FN)

Tras analizar la tabla `Students`, se ha encontrado que no cumple con la **Tercera Forma Normal (3FN)** debido a las dependencias transitivas, en las cuales el `DepartmentID` depende de una columna no clave.

### Soluci√≥n Propuesta

Para cumplir con la **Tercera Forma Normal**, proponemos los siguientes cambios:

1. **Crear una nueva tabla llamada `Departments`**:
   - Esta tabla almacenar√° la informaci√≥n relacionada con los departamentos, como el `DepartmentID` (clave primaria) y el `Name` del departamento.

2. **Modificar la tabla `Students`**:
   - Se a√±adir√° una columna `DepartmentID` que actuar√° como clave for√°nea (FK) para referenciar la tabla `Departments`.
   - De esta manera, eliminamos la dependencia transitiva, ya que la informaci√≥n del departamento se gestionar√° en una tabla separada.

### Nueva Estructura de las Tablas

<div style="text-align: center;">
    <img src="./figures/normalize_table.png" width="30%" alt="Tabla de estudiantes antes de la normalizaci√≥n" />
</div>

### Explicaci√≥n del Cambio

- **Tabla `Departments`**: Almacena la informaci√≥n √∫nica del departamento, de forma que si m√°s estudiantes pertenecen al mismo departamento, no se repita la informaci√≥n del departamento en cada fila de la tabla `Students`. Esto elimina redundancias.
  
- **Tabla `Students`**: Ahora la columna `DepartmentID` se convierte en una clave for√°nea que apunta a la tabla `Departments`. De esta forma, evitamos la dependencia transitiva entre el nombre del departamento y el resto de los atributos de los estudiantes.

Este cambio asegura que la base de datos cumpla con la **Tercera Forma Normal (3FN)** y sea m√°s eficiente en t√©rminos de almacenamiento y mantenimiento.


````python
import pandas as pd
import sqlite3

# Conectar a la base de datos SQLite
conn = sqlite3.connect("students.db")

# 1. Obtener los departamentos distintos de la tabla Students
departments = pd.read_sql_query("SELECT DISTINCT(Department) FROM Students", conn)

# 2. Crear un DataFrame con los departamentos y asignarles un ID
departments['DepartmentID'] = range(1, len(departments) + 1)

# 3. Mostrar los departamentos con su nuevo ID
print("Departamentos con ID:")
print(departments)

# 4. Crear la tabla Departments con el ID y el nombre del departamento
conn.execute("""
    CREATE TABLE IF NOT EXISTS Departments (
        DepartmentID INTEGER PRIMARY KEY,
        DepartmentName TEXT
    )
""")

# 5. Insertar los departamentos en la tabla Departments
for _, row in departments.iterrows():
    conn.execute("INSERT INTO Departments (DepartmentID, DepartmentName) VALUES (?, ?)", 
                 (row['DepartmentID'], row['Department']))

# 6. Modificar la tabla Students para agregar el campo DepartmentID (FK)
conn.execute("""
    ALTER TABLE Students 
    ADD COLUMN DepartmentID INTEGER
""")

# 7. Actualizar la tabla Students para que tenga el DepartmentID correcto
for _, row in departments.iterrows():
    conn.execute("""
        UPDATE Students 
        SET DepartmentID = ?
        WHERE Department = ?
    """, (row['DepartmentID'], row['Department']))

# Commit para guardar cambios
conn.commit()

# 8. Verificaci√≥n: Mostrar la tabla Departments
print("Tabla Departments:")
departments_from_db = pd.read_sql_query("SELECT * FROM Departments", conn)
print(departments_from_db)

# Cerrar la conexi√≥n
conn.close()
````

````python
import pandas as pd
import sqlite3

# Conectar a la base de datos SQLite
conn = sqlite3.connect("students.db")

# 1. Crear una nueva tabla sin la columna Department
conn.execute("""
    CREATE TABLE Students_temp AS
    SELECT StudentID, Name, Age, GPA, GraduationYear, DepartmentID
    FROM Students
""")

# 2. Eliminar la tabla original y renombrar la nueva
conn.execute("DROP TABLE Students")
conn.execute("ALTER TABLE Students_temp RENAME TO Students")

# 3. Verificar la estructura final
departments_from_db = pd.read_sql_query("SELECT * FROM Departments", conn)
students_from_db = pd.read_sql_query("SELECT * FROM Students", conn)

print("Tabla Departments:")
print(departments_from_db)

print("\nTabla Students con solo DepartmentID:")
print(students_from_db)

# Guardar cambios y cerrar conexi√≥n
conn.commit()
conn.close()


````

# Consultas a base normalizada
#### Explicaci√≥n de las Consultas sobre la Informaci√≥n de la Base de Datos

En esta secci√≥n, se realizaron varias consultas SQL para obtener informaci√≥n relevante de la base de datos **Students**. A continuaci√≥n, se explica cada una de las consultas realizadas:

1. **Obtener todos los estudiantes junto con el nombre de su departamento**:
    ```sql
    SELECT s.StudentID, s.Name, s.Age, s.GPA, s.GraduationYear, d.DepartmentName
    FROM Students s
    JOIN Departments d ON s.DepartmentID = d.DepartmentID LIMIT 2;
    ```
    Esta consulta utiliza una uni√≥n (`JOIN`) entre las tablas `Students` y `Departments` para obtener informaci√≥n completa de los estudiantes, incluyendo el nombre de su departamento. Se limita a mostrar solo los primeros 2 resultados.
<a id = "base_n"></a>

2. **Contar el n√∫mero de estudiantes en cada departamento**:
    ```sql
    SELECT d.DepartmentName, COUNT(*) as StudentCount
    FROM Students s
    JOIN Departments d ON s.DepartmentID = d.DepartmentID
    GROUP BY d.DepartmentName LIMIT 2;
    ```
    Esta consulta cuenta el n√∫mero de estudiantes en cada departamento. Utiliza una uni√≥n (`JOIN`) entre las tablas `Students` y `Departments` y agrupa los resultados por el nombre del departamento (`GROUP BY`). Se limita a mostrar solo los primeros 2 resultados.

3. **Obtener el promedio de edad de los estudiantes en cada departamento**:
    ```sql
    SELECT d.DepartmentName, AVG(s.Age) as AverageAge
    FROM Students s
    JOIN Departments d ON s.DepartmentID = d.DepartmentID
    GROUP BY d.DepartmentName LIMIT 2;
    ```
    Esta consulta calcula el promedio de edad de los estudiantes en cada departamento. Utiliza una uni√≥n (`JOIN`) entre las tablas `Students` y `Departments` y agrupa los resultados por el nombre del departamento (`GROUP BY`). Se limita a mostrar solo los primeros 2 resultados.

Estas consultas permiten obtener informaci√≥n detallada y resumida sobre los estudiantes y sus departamentos, facilitando el an√°lisis y la toma de decisiones dentro de la instituci√≥n educativa.

````python
import pandas as pd
import sqlite3

# Conectar a la base de datos SQLite
conn = sqlite3.connect("students.db")

# 1. Obtener todos los estudiantes junto con el nombre de su departamento
students_with_departments = pd.read_sql_query("""
    SELECT s.StudentID, s.Name, s.Age, s.GPA, s.GraduationYear, d.DepartmentName
    FROM Students s
    JOIN Departments d ON s.DepartmentID = d.DepartmentID LIMIT 2
""", conn)
print("Estudiantes con el nombre de su departamento:")
print(students_with_departments)

# 2. Contar el n√∫mero de estudiantes en cada departamento
student_count_per_department = pd.read_sql_query("""
    SELECT d.DepartmentName, COUNT(*) as StudentCount
    FROM Students s
    JOIN Departments d ON s.DepartmentID = d.DepartmentID
    GROUP BY d.DepartmentName LIMIT 2
""", conn)
print("\nN√∫mero de estudiantes por departamento:")
print(student_count_per_department)



# 3. Obtener el promedio de edad de los estudiantes en cada departamento
average_age_per_department = pd.read_sql_query("""
    SELECT d.DepartmentName, AVG(s.Age) as AverageAge
    FROM Students s
    JOIN Departments d ON s.DepartmentID = d.DepartmentID
    GROUP BY d.DepartmentName
    LIMIT 2
""", conn)
print("\nPromedio de edad por departamento:")
print(average_age_per_department)

# Cerrar la conexi√≥n
conn.close()
````
